```python
#Stacked bar plot to show the contribution of each sector to the demand

regions = (
    "Lowlands",
    "Valleys",
    "Highlands",
)
dimensions = {
    "Shelter and thermal comfort": np.array([97, 5, 4.3]),
    "Nutrition": np.array([73, 64, 61]),
    "Hygiene": np.array([64, 109, 101]),
    "ICT": np.array([8, 8, 8]),
}
width = 0.5

fig, ax = plt.subplots(figsize=[5,10])
bottom = np.zeros(3)

colors = {
    "Shelter and thermal comfort": "gold",
    "Nutrition": "blue",
    "Hygiene": "green",
    "ICT": "orange",
}

for boolean, dimensions in dimensions.items():
    p = ax.bar(regions, dimensions, width, label=boolean, bottom=bottom, color=colors[boolean], alpha=.5)
    bottom += dimensions

ax.set_title("Contribution of each DLS dimention to electricity demand ")
ax.set_ylabel("Monthly electricity demand (kWh)") 
ax.legend(loc="upper right")

plt.savefig('contribution_dls_monthly', dpi=600, bbox_inches='tight')

plt.show()
```
